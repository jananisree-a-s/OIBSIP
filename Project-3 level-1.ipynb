```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("AB_NYC_2019.csv")

# Preview data
data.head(10)
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data['name'] = data['name'].fillna("Unknown")
data['host_name'] = data['host_name'].fillna("Unknown")
data['reviews_per_month'] = data['reviews_per_month'].fillna(0)
data['availability_365'] = data['availability_365'].fillna(data['availability_365'].median())
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data['neighbourhood_group'] = data['neighbourhood_group'].str.lower()
data['room_type'] = data['room_type'].str.lower()
data = data[data['price'] > 0]
```

```python
plt.figure()
sns.boxplot(x=data['price'])
plt.title("Price Outliers")
plt.show()
```

```python
upper_limit = data['price'].quantile(0.99)
data = data[data['price'] <= upper_limit]
```

```python
data.describe()
```

```python
data.info()
```

```python
data.to_csv("AB_NYC_2019_CLEANED.csv", index=False)
```
